## **Desafio Prático: Lançamento de Moedas**


Você tem uma moeda justa (probabilidade de Cara = 0.5, Coroa = 0.5).

Implemente uma simulação simples do lançamento de uma moeda para observar a Lei dos Grandes Números e calcular probabilidades básicas.

Passos Sugeridos:

**1.   Simulação de Lançamentos**

* Simule o lançamento de uma moeda 10, 100, 1000 e 10.000 vezes.

* Para cada lançamento, registre o resultado (Cara ou Coroa).

**2.   Lei dos Grandes Números**

* Para cada série de lançamentos (10, 100, 1000, 10.000), calcule a proporção de vezes que a moeda caiu "Cara".

* Plote um gráfico mostrando como a proporção de "Caras" se aproxima da probabilidade teórica (0.5) à medida que o número de lançamentos aumenta.

**3.   Probabilidade Observada**

* Calcule a probabilidade observada de obter "Cara" para o maior número de lançamentos (10.000).

* Compare este valor com a probabilidade teórica de 0.5.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import norm
```

```python
def simular_lancamento_moeda(cara_coroa):
    resultados = np.random.randint(1, 3, cara_coroa)
    medias_acumuladas = np.cumsum(resultados) / np.arange(1, cara_coroa + 1)
    return medias_acumuladas

# Simulações para diferentes números de lançamentos
cara_coroa_list = [10, 100, 1000, 10000]

plt.figure(figsize=(12, 8))
for cara_coroa in cara_coroa_list:
    medias = simular_lancamento_moeda(cara_coroa)
    plt.plot(range(1, cara_coroa + 1), medias, label=f'{cara_coroa} lançamentos')

plt.axhline(y=3.5, color='r', linestyle='--', label='Média Teórica (3.5)')
plt.xlabel('Número de Lançamentos')
plt.ylabel('Média Observada')
plt.title('Convergência da Média Observada para a Média Teórica (Lei dos Grandes Números)')
plt.legend()
plt.grid(True)
plt.show()
```

## **Desafio Prático: Controle de Qualidade e Manutenção Preditiva**

Uma fábrica produz componentes eletrônicos. Historicamente, 5% dos componentes produzidos são defeituosos. A fábrica implementou um novo sensor que detecta uma anomalia em componentes. Sabe-se que:



*   O sensor detecta corretamente 90% dos componentes defeituosos (sensibilidade).
*   O sensor, erroneamente, aponta anomalia em 8% dos componentes não defeituosos (falso positivo).
* O peso dos componentes segue uma distribuição não-normal (ex: uniforme) entre 10g e 12g.


Utilize simulações em Python para consolidar os conceitos de Lei dos Grandes Números (LGN), Teorema Central do Limite (TCL) e Teorema de Bayes.

Passos Sugeridos:

1. **Lei dos Grandes Números - Estimando a Taxa de Defeitos**

* Simule a produção de 10, 100, 1000 e 10.000 componentes.

* Atribua o status de "defeituoso" ou "não defeituoso" com base na taxa histórica de 5%.

* Plote a convergência da taxa observada para a taxa teórica de 5%.

2. **Teorema Central do Limite - Análise do Peso**

* Gere uma população de 100.000 pesos de componentes com distribuição uniforme (10g a 12g).

* Extraia 1000 amostras de tamanho n=40 e calcule a média de peso de cada amostra.

* Plote o histograma das médias amostrais e observe a formação da curva Normal.

3. **Teorema de Bayes - Confiabilidade do Sensor**

* Defina as probabilidades a priori e condicionais fornecidas no cenário.

* Calcule a probabilidade de um componente ser realmente defeituoso dado que o sensor detectou uma anomalia: P(Defeituoso | Anomalia).

* Compare o resultado com a intuição inicial.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import norm
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)

tamanhos = [10, 100, 1000, 10000]
for n in tamanhos:
    defeituosos = np.random.random(n) < 0.05
    print(f"n = {n}: taxa observada = {defeituosos.mean():.4f}")

n_max = 10000
defeitos = (np.random.random(n_max) < 0.05).astype(int)
taxa_acumulada = np.cumsum(defeitos) / np.arange(1, n_max + 1)

plt.figure(figsize=(10, 4))
plt.plot(taxa_acumulada)
plt.axhline(y=0.05, color='r', linestyle='--')
plt.xlabel('Número de componentes')
plt.ylabel('Taxa observada de defeitos')
plt.title('Lei dos Grandes Números')
plt.show()

populacao = np.random.uniform(10, 12, 100000)
medias = [np.random.choice(populacao, 40).mean() for _ in range(1000)]

plt.figure(figsize=(10, 4))
plt.hist(medias, bins=30, density=True, alpha=0.7)
x = np.linspace(min(medias), max(medias), 100)
plt.plot(x, stats.norm.pdf(x, np.mean(medias), np.std(medias)), 'r-')
plt.title('Teorema Central do Limite - Médias Amostrais (n=40)')
plt.show()

P_defeito = 0.05
P_anomalia_dado_defeito = 0.90
P_anomalia_dado_bom = 0.08

P_anomalia = P_anomalia_dado_defeito * P_defeito + P_anomalia_dado_bom * 0.95
P_defeito_dado_anomalia = (P_anomalia_dado_defeito * P_defeito) / P_anomalia

print(f"\nP(Defeituoso | Anomalia) = {P_defeito_dado_anomalia:.2%}")
```
